```python
from btree import *
from anytree import Node, RenderTree
from anytree.exporter import UniqueDotExporter
from IPython.display import Image, display
```

# Graphics

```python

def make_anytree(b_tree: BTreeNode):
    root = Node(b_tree.keys, children=[make_anytree(c) for c in b_tree.children])
    return root

def display_anytree(anytree:Node):
    UniqueDotExporter(anytree).to_picture("/tmp/anytree.png")
    display(Image("/tmp/anytree.png"))
```

```python
tree = BTree(2)
keys = [59, 23, 7, 97, 73, 67, 19, 79, 61, 41, 42, 69, 98, 12, 14, 20, 21, 25, 1, 2, 3, 4, 5, 44, 45, 46, 47, 48, 49, 50]

print("Inserting...")
for key in keys:
    tree.insert(key)
    display_anytree(make_anytree(tree.root))

print("Deleting...")
for key in keys:
    tree.delete(key)
    display_anytree(make_anytree(tree.root))

```
