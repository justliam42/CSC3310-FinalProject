# Graphics

```python
def format_lists(node):
    lists_str = "\\n".join(f"- {item}" for item in node.lists_attr)
    return f'label="{node.name}\\n{lists_str}"'

def node_attrs(node):
    attrs = ['shape=box', 'style=filled', 'fillcolor=lightyellow', 'fixedsize=False']
    attrs.append(format_lists(node))
    return ', '.join(attrs)

def make_anytree(b_tree: BTreeNode):
    root = Node(', '.join(str(x) for x in b_tree.keys).strip(', '), children=[make_anytree(c) for c in b_tree.children])
    return root

def display_anytree(anytree:Node):
    UniqueDotExporter(
    anytree,
    ).to_picture("/tmp/anytree.png")
    display(Image("/tmp/anytree.png"))
```

```python
tree = BTree(2)
keys = [59, 23, 7, 97, 73, 67, 19, 79, 61, 41, 42, 69, 98, 12, 14, 20, 21, 25, 1, 2, 3, 4, 5, 44, 45, 46, 47, 48, 49, 50]

print("Inserting...")
for key in keys:
    tree.insert(key)
    print("insert:", key)
    display_anytree(make_anytree(tree.root))

print("Deleting...")
for key in keys:
    tree.delete(key)
    print("delete:",key)
    display_anytree(make_anytree(tree.root))

```
